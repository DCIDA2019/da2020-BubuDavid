## Operaciones con más de una tabla

El objetivo de esta "subsección" es intentar, juntar 3 tablas y observar las temperaturas correspondientes a los estados correctos.

```python
# Definiendo algunas tablas, lluvia, temperatura media y temperatura máxima
tempMin = pd.read_csv("Actividad2.1/TemperaturaMinima2019.csv", encoding="Latin-1")
tempMedia = pd.read_csv("Actividad2.1/TemperaturaMedia2019.csv", encoding="Latin-1")
tempMax = pd.read_csv("Actividad2.1/TemperaturaMaxima2019.csv", encoding="Latin-1")
```

Podemos sumarlas:

```python
tempMax + tempMedia + tempMin
```

Notar que la tabla anterior tiene una estructura muy extraña, no es para nada lo que buscamos, en realidad, lo que esperamos es juntar las Tmax, Tmed y Tmin de tal manera que tenga sentido lo que estamos viendo. Hay que analizar las tablas y ver si tienen algo en común en lo que nos podamos bassar para juntarlas

```python
tempMax
```

```python
tempMin
```

```python
tempMedia
```

*Tal vez podemos usar concatenate de pandas*

```python
pd.concat([tempMin, tempMedia, tempMax])
```

Esto va un paso mejor, lo que pasa es que no nos está "mergeando" las columnas, sólo nos está pegando una tabla con otra (Natural con el nombre de la función concat)

*NOTA SALVAJE:* Lo mismo que hicimos anteriormente, se puede hacer de una manera más corta haciendo:

```python
tempMin.append([tempMedia, tempMax])
```

Volviendo al asunto de hacer "Merge" de las tablas, podemos justamente intentar con esta función... con Merge, sin embargo eso lo haremos en otro archivo, ya que eso es parte de la actividad 2 parte 3.

*También podemos renombrar las columnas de la siguiente manera*

```python
tempMin = tempMin.rename(columns={"Edo": "Estado"}) 
```

Y reordenar columnas de la siguiente manera (Esto nos servirá para la siguiente sección de la tarea)

```python
cols = ['Estado']  + [col for col in tempMin if col != 'Estado']
tempMin = tempMin[cols]
```

```python
tempMin
```

#### En este último caso, cambiamos el nombre de Edo por Estado y lo posicionamos hasta adelante :D
