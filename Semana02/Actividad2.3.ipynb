```python
# Importando cosas chulas
import pandas as pd
import numpy as np
```

### Temperaturas

```python
# Definiendo algunas tablas, temperatura mínima, temperatura media y temperatura máxima
tempMin = pd.read_csv("Actividad2.1/TemperaturaMinima2019.csv", encoding="Latin-1")
tempMedia = pd.read_csv("Actividad2.1/TemperaturaMedia2019.csv", encoding="Latin-1")
tempMax = pd.read_csv("Actividad2.1/TemperaturaMaxima2019.csv", encoding="Latin-1")
```

*Como vimos en la sección pasada, podemos sumar las tablas, con riesgo de que no salga lo que esperamos, también se pueden hacer operaciones entre columnas o filas, sin embargo, lo que usaremos en este caso es el método "merge" de pandas, que nos permite juntar varias tablas, sin perder información y uniendo las filas y columnas que son iguales en varias tablas, de la siguiente manera:*

```python
mergedTemperature = pd.merge(tempMin, tempMedia)
mergedTemperature = pd.merge(mergedTemperature, tempMax)
```

Veamos cuáles son las diferencias entre las 3 tablas por separado y las tabla "merge", esto con ayuda de un objeto que crearemos a continuación llamada *Display*

```python
class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

```python
display('tempMin', 'tempMedia', 'tempMax', 'mergedTemperature')
```

A pesar de que únicamente podemos ver algunas filas, se nota claramente la diferencia a un simple append. La tabla llamada *mergedTemperature* es la tabla obtenida por unir todas las tablas de temperaturas, ahora haremos uno para las lluvias.

### Lluvias

```python
lluvias = pd.read_csv("Actividad2.1/Lluvias2019.csv", encoding="Latin-1")
```

```python
lluvias
```

*Ahora viene la parte complicada, juntar la tabla de lluvias con la tabla de temperaturas, esto es especialmente complicado ya que debemos notar que no tienen los mismos nombres de columna, por lo que si hacemos un merge, veamos lo que pasa* 

```python
pd.merge(mergedTemperature, lluvias)
```

Esto es un error, ya que no coincide en lo absoluto, ninguna columna, aunque los datos de latitud, longitud y estado son los mismos, los nombres de las columnas no lo son, los de lluvias están en mayúsculas y los de la tabla merge no

*Lo que tenemos que hacer es, renombrar las columnas de la tabla de lluvias, además de remover algunas columnas que no nos sirven para juntar las tablas, se pierde información, sin embargo no se puede hacer de otra manera... Bueno, sí, creando un diccionario a mano de más de 600 registros, cosa que por supuesto, no haré*

```python
lluvias.rename(columns=lambda x: x.capitalize(), inplace=True)
lluvias = lluvias.drop(columns="Nombre")
lluvias = lluvias.drop(columns=["Lon", "Lat"])
lluvias = lluvias.rename(columns={"Clave_sih":"Clave"})
```

```python
lluvias
```

Una vez cambiados los nombres, podemos hacer merge

```python
completeTable = pd.merge(mergedTemperature, lluvias)
completeTable
```

Ahora, sólo para presumir que sé cambiar columnas de lugar, moveremos la "Clave" al principio

```python
cols = ['Clave']  + [col for col in completeTable if col != 'Clave']
completeTable = completeTable[cols]
```

Y por último, tenemos por fin lo que queríamos, le cambiamos el nombre a "Est" por ciudad, porque de hecho eso es lo que representa

```python
completeTable = completeTable.rename(columns={"Est":"Ciudad"})
```

```python
completeTable
```

### FINALLY! Dio resultado, juntamos la tabla de lluvias con el de temperaturas y todos felices y contentos...
Note que no están todos los registros porque algunos registros son completamente disintos de una tabla a otra, aunque se trate de la misma ciudad. La manera en cómo escribieron los datos, está bien fea

Ya por último: Exportaremos el resultado final a un archivo csv como vimos en la actividad 2.2

```python
completeTable.to_csv("Tabla de lluvias y temperaturas unidas.csv")
```
