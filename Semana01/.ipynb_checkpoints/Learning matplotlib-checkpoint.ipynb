```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

# Learning matplotlib

```python
x = np.arange(-5, 5, 0.5)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, x**2, 'r--', label=r"$x^2$")
plt.plot(x, x**3, 'k', label=r"$x^3$")

plt.xlabel("No sé que está pasando en x")
plt.ylabel("No sé que está pasando en Y")
plt.title("No sé que está pasando ayuda")
plt.savefig("Hello world-matplotlib.pdf")
```
