```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Rain and Temperatures 

```python
# To print extended table
#pd.set_option('display.max_columns', None)  # or 1000
#pd.set_option('display.max_rows', None)  # or 1000
```

```python
# Importing table
data = pd.read_csv("../../Semana2/MergedRainTemperature.csv", usecols=list(range(1,10)))
```

```python
# Sorting table
data = data.sort_values(by="Edo")
data
```

```python
# Use the scatter function
fig, ax = plt.subplots(figsize=(10, 8), constrained_layout=True)

ax.grid()
ax.scatter(data["Edo"], data["Tmin"], alpha=0.5, color="blue")
ax.scatter(data["Edo"], data["Tmed"], alpha=0.5, color="orange")
ax.scatter(data["Edo"], data["Tmax"], alpha=0.5, color="red")
plt.setp(ax.get_xticklabels(), rotation=70, ha="right")
plt.show()
```

```python
# Setting an index for average
data = data.set_index(["Edo", "Clave"])
data
```

```python
# Calculing the average
data_avg = data.mean(level='Edo')
# This below line works for graph
data_avg = data_avg.reset_index().sort_values("Tmed")
data = data.reset_index()
```

```python
# Ploting the average thingys
fig, ax = plt.subplots(figsize=(10, 8), constrained_layout=True)

ax.grid()
ax.plot(data_avg["Edo"], data_avg["Tmin"], alpha=0.7, color="blue")
ax.plot(data_avg["Edo"], data_avg["Tmed"], alpha=0.7, color="orange")
ax.plot(data_avg["Edo"], data_avg["Tmax"], alpha=0.7, color="red")
plt.setp(ax.get_xticklabels(), rotation=70, ha="right")
plt.show()
```

```python
# Now merge both graphs
fig, ax = plt.subplots(figsize=(10, 8), constrained_layout=True)

ax.grid()
ax.plot(data_avg["Edo"], data_avg["Tmin"], alpha=0.8, color="blue")
ax.plot(data_avg["Edo"], data_avg["Tmed"], alpha=0.8, color="orange")
ax.plot(data_avg["Edo"], data_avg["Tmax"], alpha=0.8, color="red")
ax.scatter(data["Edo"], data["Tmin"], alpha=0.1, color="blue")
ax.scatter(data["Edo"], data["Tmed"], alpha=0.1, color="orange")
ax.scatter(data["Edo"], data["Tmax"], alpha=0.1, color="red")
plt.setp(ax.get_xticklabels(), rotation=70, ha="right")
plt.show()
```

```python
data
```

```python
plt.scatter(data['Lon'], data['Lat'], color="Purple", s=data["Lluvias"]/2, alpha=0.5)
plt.xlabel('Latitud')
plt.title("Estados con mayor precipitación")
plt.ylabel('Longitud')
plt.show()
```

```python
import matplotlib.pyplot as plt

N = 9
fig, ax = plt.subplots(figsize=(20, 10))
ind = np.arange(N)    # the x locations for the groups
width = 0.2        # the width of the bars

groupA = tuple(salData["A"][36:])
groupB = tuple(salData["B"][36:])
groupC = tuple(salData["C"][36:])

p1 = ax.bar(ind, groupA, width, bottom=0, color="#94d3ac")
p2 = ax.bar(ind + width, groupB, width, bottom=0, color="#ffd271")
p3 = ax.bar(ind + 2*width, groupC, width, bottom=0, color="#e78fb3")

ax.set_title('Salarios mínimos dependiendo del tiempo (de 1991 a 1998)')
ax.set_xticklabels(list(salData["Fecha de publicación el DOF"][36:])[-1:0:-1])
plt.setp(ax.get_xticklabels(), rotation=70, ha="left")

ax.legend((p1[0], p2[0], p3[0]), ('A', 'B', 'C'))
ax.autoscale_view()

plt.show()
```
